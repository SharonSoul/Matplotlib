```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,3,5]
y = [6,9,11]
```

```python
plt.plot(x, y)
```

```python
plt.scatter(x, y)
```

```python
# The line can be modified here, shorthand method is (color)()
plt.plot(x, y, label = "Amount Sold", color = "Red", marker='*', linewidth = 2, linestyle = "dashed", markersize = 14)

# USe a title with customised fonts and colours
plt.title("FIRST EVER GRAPH", fontdict={"color": "maroon", "fontsize": 20, "fontweight": 1000, "fontname": "Cursive"})

# Use a title for the axis with customised fonts and colors
plt.xlabel("X axis", fontdict = {"color": "blue", "fontsize": 20, "fontname": "Calibri"})
plt.ylabel("Y axis", fontdict = {"color": "blue", "fontsize": 20})

# USe specified starting and ending points for the counts on either axis
plt.xticks([1, 2, 3, 4, 5])
plt.yticks([5, 7, 9, 11, 13])

# To show a legend, a label must be inside the plot function/method. bbox_to_anchor brings the legend outside the graph
plt.legend(bbox_to_anchor = (1.35, 1), title = "Legend")

# whether to show the plt comments or not
plt.show()
```

```python
a = [1, 2, 4, 4.5, 7]
b = [3, 6, 9, 11, 14]


plt.title("SECOND BEST GRAPH!", fontdict = {"color": "Green", "fontweight": 1000, "fontsize": 20, "fontname": "Cursive"})

plt.plot(a, b, "b*-.", label = "Amount Scored")
plt.xlabel("X axis", fontdict = {"color": "lawngreen", "fontsize": 30, "fontname": "gabriola"})
plt.ylabel("Y axis", fontdict = {"color": "lawngreen", "fontsize": 30, "fontname": "gabriola", "fontweight": 1000})
plt.legend(title = "Scores", bbox_to_anchor = (1.4, 1))

plt.show()
```

```python

x1 = np.arange(0, 4.5, 0.5)
x1
```

```python
x2 = x1**2
x2
```

```python
# Printing more than one line on a graph

plt.plot(x1, x2, "b*-.")
plt.plot(x, y, "g*-.")
plt.plot(a, b, "r*-.")
plt.figure(figsize = (50, 5), dpi = 700)

plt.savefig("My First Graph.png")
plt.show()
```

```python
# To show a prediction on a line with two 

plt.plot(x1[:5], x2[:5] ** 2, "b^--")
plt.plot(x1[4:], x2[4:] ** 2, "r*-.")

plt.show()
```

# Creating a Bar Chart with Matplotlib


```python
labels = ["Sunday", "Monday", "Tuesday"]
values = [200, 600, 150]
```

```python
plt.figure(figsize = (10,7))

#bars = plt.bar(labels, values, width = 0.9)
plt.title("My First Bar Chart", fontdict ={"color": "lawngreen", "fontsize" : 30, })
bars = plt.bar(labels, values, width = 0.9, align = "center", color = "Green", edgecolor = "lawngreen", linewidth = 3)
plt.xticks(rotation = 90)
bars[0].set_hatch("*")
bars[1].set_hatch("+")
bars[2].set_hatch("/")

plt.show()
```
